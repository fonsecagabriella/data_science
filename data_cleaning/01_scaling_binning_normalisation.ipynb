```python
import pandas as pd

df = pd.DataFrame(data['facilities'])
df
```

```python
# Step 1: Calculate total violations
df['total_violations'] = df['critical_violations'] + df['non_critical_violations']
df
```

```python
# Step 2: Scale facility size to per 1,000 sqft
df['scaled_size'] = df['square_footage'] / 1000
df
```

```python
# Step 3: Scale violations
df['violations_per_1000_sqft'] = round(df['total_violations'] / df['scaled_size'], 2)
df
```

```python
# Quantile binning 
# this functions creates a new category in your date, by splitting them in evenly sizes

df['volume_category_quantiles'] = pd.qcut(df['production_volume'], q=3, labels=['Low', 'Moderate', 'High'], duplicates='drop')
df
```

```python
df['volume_category_quantiles'].value_counts()
```

```python
# Percentile binning
# here we define the values manually, taking into account the data distribution
volume_bins = [
	df['employee_count'].min() - 1,  # Start at the minimum value
	df['employee_count'].quantile(0.33),  
	df['employee_count'].quantile(0.66),  
	df['employee_count'].max() + 1  # End at the maximum value
]

df['employee_count_percentiles'] = pd.cut(df['employee_count'], bins=volume_bins, labels=['Low', 'Medium', 'High'])
df
```

```python
df['employee_count_percentiles'].value_counts()
```

```python
# Function to normalize values to a 0-1 range
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

# Normalize "Days Since Last Inspection" (fewer days = better, so we invert it)
df["days_score"] = 1 - normalize(df["days_since_last_inspection"])

# Normalize "Violations per 1,000 sqft" (fewer violations = better, so we invert it)
df["violations_score"] = 1 - normalize(df["violations_per_1000_sqft"])

# Compute "Production Volume per Employee"
df["production_volume_per_employee"] = (df["production_volume"] / (df["employee_count"]))

# Normalize "Production Volume per Employee" (lower is better)
df["production_volume_per_employee_score"] = 1 - normalize(df["production_volume_per_employee"])

# Calculate the composite score using weighted factors
df["composite_score"] = round(
    (df["days_score"] * 0.30 + df["violations_score"] * 0.50 + df["production_volume_per_employee_score"] * 0.2) * 100
)
```
