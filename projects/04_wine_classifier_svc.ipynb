# Wine Classifier with SVC Example

```python
# load ds
from sklearn.datasets import load_wine

wine = load_wine(as_frame=True)

print(wine.DESCR)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, random_state=42)

X_train.head(10)
```

```python
y_train.head(10)
```

```python
# with SVC start with scaling the features

# starting with a simple linear SVC
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import  StandardScaler

linear_svc = make_pipeline(StandardScaler(), 
LinearSVC(dual=True, random_state=42 ))

linear_svc.fit(X_train, y_train)
```

```python
# check the performance of the model
from sklearn.model_selection import cross_val_score

cross_val_score(linear_svc, X_train,y_train).mean()
```

```python
# implementing a kernel SVM with parameter tunning
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform, uniform
from sklearn.svm import SVC

svm_classifier =  make_pipeline(StandardScaler(), SVC(random_state=42))

parameter_grid = {"svc__gamma": loguniform(0.001,0.1),
"svc__C": uniform(1, 10)}

svm_search = RandomizedSearchCV(svm_classifier, parameter_grid, n_iter=100, cv=5, random_state=42)

svm_search.fit(X_train, y_train)
```

```python
svm_search.best_score_
```

```python
svm_search.score(X_test, y_test)
```
