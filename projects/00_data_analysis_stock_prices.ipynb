```python
# relationship between the prices of AAPL and GOOGL stock
# difference between the close & open prices - how much did the stock gain or lose that day

# select the AAPL & GOOGL stocks
columns = ["AAPL", "GOOGL"]
aapl_googl = df[ df["ticker"].isin(columns)]
```

```python
aapl_googl.sample(10)
```

```python
# create columns for the AAPL & GOOGL differences per day
aapl_googl["AAPL_diff"] = aapl_googl[ aapl_googl["ticker"] == "AAPL"]["close"] - aapl_googl[ aapl_googl["ticker"] == "AAPL"]["open"]
```

```python
aapl_googl.sample(10)
```

```python
aapl_googl["GOOGL_diff"] = aapl_googl[ aapl_googl["ticker"] == "GOOGL"]["close"] - aapl_googl[ aapl_googl["ticker"] == "GOOGL"]["open"]
```

```python
# plotting with scatterplot
ax = aapl_googl.plot(kind="scatter", x="AAPL_diff", y="GOOGL_diff", alpha=0.3)
plt.title("Relationship between AAPL and GOOGL stock gains per day", fontsize=16)
plt.xlabel("AAPL close - open price", fontsize=12)
plt.ylabel("GOOGL close - open price", fontsize=12)

plt.grid()
```

```python
# import the ticker tools
import matplotlib.ticker as ticker

# plotting with scatterplot
ax = aapl_googl.plot(kind="scatter", x="AAPL_diff", y="GOOGL_diff", alpha=0.3)
plt.title("Relationship between AAPL and GOOGL stock gains per day", fontsize=16)
plt.xlabel("AAPL close - open price", fontsize=12)
plt.ylabel("GOOGL close - open price", fontsize=12)

ax.xaxis.set_major_locator(ticker.MultipleLocator(2))

plt.grid()
```

```python
# plotting with scatterplot
ax = aapl_googl.plot(kind="scatter", x="AAPL_diff", y="GOOGL_diff", color="purple", marker="x", alpha=0.3)
plt.title("Relationship between AAPL and GOOGL stock gains per day", fontsize=16)
plt.xlabel("AAPL close - open price", fontsize=12)
plt.ylabel("GOOGL close - open price", fontsize=12)

ax.xaxis.set_major_locator(ticker.MultipleLocator(2))

plt.xlim(-8,14)
plt.ylim(-8,8)

plt.grid(alpha=0.5)
```

```python
# average trading volume per year, compare within a stock - [AMZN, META, DIS]

# select all three columns
filtered_df = df[ df["ticker"].isin(["AMZN", "META", "DIS"])]
filtered_df.sample(10)
```

```python
# group by each ticker and year
grouped_by_ticker_and_year = filtered_df.groupby(["ticker", filtered_df.index.year])
grouped_by_ticker_and_year
```

```python
# aggregate by calculating the mean of the volume feature
average_volume = grouped_by_ticker_and_year["volume"].mean()
average_volume.sample(10)
type(average_volume)
```

```python
average_volume.plot(kind="bar")
```

```python
average_volume.unstack().plot(kind="bar")
```

```python
ax = average_volume.unstack().plot(kind="bar")
plt.title("Average trading volume per year per stock", fontsize=16)
plt.xlabel("Ticker", fontsize=12)
plt.ylabel("Average trading volume per year", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Year", bbox_to_anchor=(1.05,1))
```

```python
pink_shades = [
    "#FFC0CB",  # Light Pink
    "#FFB6C1",  # Slightly Darker Light Pink
    "#FF69B4",  # Hot Pink
    "#FF1493",  # Deep Pink
    "#DB7093",  # Pale Violet Red
    "#C71585",  # Medium Violet Red
    "#B03060",  # Rich Maroon Pink
    "#A020F0",  # Dark Orchid-like Pink
    "#800080",  # Purple-ish Pink
    "#660033",  # Very Dark Pink
]

ax = average_volume.unstack().plot(kind="bar", color=pink_shades)
plt.title("Average trading volume per year per stock", fontsize=16)
plt.xlabel("Ticker", fontsize=12)
plt.ylabel("Average trading volume per year", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Year", bbox_to_anchor=(1.05,1))
```
