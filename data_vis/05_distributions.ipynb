```python
# Environment setup
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

filepath = "../data/iris.csv"
iris = pd.read_csv(filepath, index_col="Id")

iris.head()
```

```python
# we can creta a histogram to see how petal lenght varies in iris flowers

sns.histplot(iris["Petal Length (cm)"])
```

```python
# KDE = Kernel Density Estimate 
# it's basically a smooth histogram

sns.kdeplot(data=iris["Petal Length (cm)"], shade=True) # the parameter 'shade' colours the area below the curve
```

```python
# 2D KDE plots shows the combinations of two different attributes
# the darker part of the figure are more likely to appear (when use kind=kde)

sns.jointplot(data=iris, x="Petal Length (cm)", y="Sepal Width (cm)", kind="kde", shade=True )
```

```python
# histograms for each specieis
sns.histplot(data=iris, x="Petal Length (cm)", hue="Species")

plt.title("Histogram of Petal Lenghts, by Species")
```

```python
sns.kdeplot(data=iris, x="Petal Length (cm)", hue="Species", shade="True")
```
