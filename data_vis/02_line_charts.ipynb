```python
# Environment setup
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

print("Setup complete")
```

```python
# Load dataset and check first 5 rows

file_path = "../data/spotify.csv"

spotify_data = pd.read_csv(file_path, index_col=0, parse_dates=True)

spotify_data.head()
```

```python
# last 5 rows
spotify_data.tail()
```

```python
# plot the data -  simple
sns.lineplot(spotify_data)
```

```python
# plot the data, with some modifications in size and title for the graph

# set the width and height of the figure
plt.figure(figsize=(8, 8))

# add title
plt.title("Daily  Global Streams of Popular Songs in 2017-2018")

# Line chart
sns.lineplot(spotify_data)
```

```python
# plotting subset of columns

plt.figure(figsize=(14,6))
plt.title("Daily Streams from 'Shape of You' vs 'Despacito'")

# line chart showing 'Shape of you'
sns.lineplot(data=spotify_data["Shape of You"], label="Ed Sheeran, Shape of You")

# line chart showing 'Despacito'
sns.lineplot(data=spotify_data["Despacito"], label="I have no idea who, Despacito")

# Add label for horizontal axis
plt.xlabel("Date")
```
