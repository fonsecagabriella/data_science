## Concepts

**Timeseries data**
Data that is equally spaced in time / indexed in time, with equal intervals.

**Seasonality**
A pattern that repeats over fixed intervals of time.

**Trend**
General direction of the series.

**Residuals**
Changes in the pattern not explained by the seasonality and trend, ideally they are random.

‚úÖ **NOTE:** **We can model the seasonality and trend** but not the residuals, so a model is never perfect.

## Baseline Model

- A good baseline model is very important in forecasting. In this tutorial we use (statsforecast) to create a baseline model. For season data, usually **SeasonalNaive** model is the best one.

## Statistical models for forecasting

- ARIMA (AutoRegressive Integrated Moving Average) model
- SARIMA (seasonal ARIMA model)
- Exponential smoothing
- Theta model (decomposition approach)
- MSTL, BATS, TBATS (multiple seasonal periods)
- Croston, IMAPA, TSB (intermittent series - data with zero and non-zero values)

### **ARIMA**
With ARIMA you can model and forecast any series with a trend and one seasonal period.

- **ARL** / ***autoregressive model***: future values of the series depend on past values. Denoted as AR(p).
- **MA** /  ***moving average model***: Future values depend on the present and past error terms; the errors are independent and normally distributed.


## Forecasting with exogenous features

External features that are no part of the timeseries itself (holidays, temperature outside, wind speed, etc)

**Types**
- static (stays constant in time, like a product category)
- historical (we know the value in the past but not in the future, like price of gas)
- future features (values that known both in past and future, like holidays)

**Create exogenous features** 
- Use Fourier terms for seasonality
- Enconde timestamp as features (code then, like mondays = day 1, which will repeat every week)

***‚ÄºÔ∏è NOTE: Not all models support exogenous features, but ARIMA/SARIMA does.***

## Prediction intervals

- A point forecast does not consider the uncertainty of the future. 
- Stochastic models generate uncertainty automatically following the normal distribution (ARIMA is a a stochastic model)
