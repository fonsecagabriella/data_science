# Pandas Basics

```python
# convert to datetime format
import pandas as pd

stock_prices = pd.read_csv("../data/stock_prices.csv")


stock_prices["datetime"] = pd.to_datetime(stock_prices["date"])

# additional methods
print(stock_prices["datetime"][0].weekday())

# attributes
print(stock_prices["datetime"][0].month)

# to access a series of attributes, you need to use dt
# example, create a column for month
stock_prices["month"] = stock_prices["datetime"].dt.month

# set datetime as df index in the current df (by using inplace=True)
stock_prices.set_index("datetime", inplace=True)

# select a single row with a certain date
stock_prices.loc["2024-04-16"]
```

```python
#remove Unnamed column (original index)
stock_prices.drop("Unnamed: 0", axis=1, inplace=True)

# analyse trends over time
stock_prices.describe()
```

## Simple plotting + Moving averages

```python
# plot timeseries data where index is a datetime
stock_prices["volume"].plot(kind="line")
```

Moving averages help you to smooth the data, so you can focus on bigger patterns instead of noise in the data.

```python
# rolling takes the moving average of a period defined by the value passed by window when used with mean()
# ps: the first X values of the df won't have the moving average, it will be defined as NaN
stock_prices["volume"].rolling(window=90).mean().plot(kind="line")
```

```python
# calculate percentage of change
stock_prices["percent_change"] = stock_prices["close"].pct_change()

# the first value will be NaN as there's nothing to compare it to
print(stock_prices["percent_change"].head(20))
```

## Grouping & Segmentation

```python
# Grouping

nike_stocks = stock_prices[stock_prices["ticker"]=="NKE"]

nike_stocks.sample(10)

nike_stocks["close"].plot(kind="line") 
```

```python
# grouping by year
nike_stocks.groupby(nike_stocks.index.year)["close"].describe()
```

```python
# grouping by year
nike_stocks.groupby(nike_stocks.index.year)["close"].mean().plot(kind="line")
```

```python
# grouping, select the column you want, apply the mean, and sort the values
stock_prices.groupby("sector")["close"].mean().sort_values()
```

## Ploting multiple lines in a graph

```python
# plotting multiple values
nike_stocks.plot(y=["high", "low"])
```

```python
nike_stocks.groupby(nike_stocks.index.year)[["close", "low"]].mean().plot(kind="line")
```
