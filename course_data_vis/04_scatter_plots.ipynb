```python
# Environment setup
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

insurance_data = pd.read_csv("../data/insurance_claims.csv")

insurance_data.head()

```

```python
plt.figure(figsize=(15, 6))


# to plot 3 figures
sns.scatterplot(y=insurance_data["total_claim_amount"], x=insurance_data["age"], hue=insurance_data["fraud_reported"])

# note, another way to write the above
# sns.scatterplot(data=insurance_data, y="total_claim_amount", x="age", hue="fraud_reported")


plt.xlabel("Age")
plt.ylabel("Total Claim Amount")
plt.title("Months as Customer vs Amount Claimed")




# adds a regression line
# Note: if you use the regline, the hue doesn't work
# sns.regplot(y=insurance_data["total_claim_amount"], x=insurance_data["age"])


plt.show()


```

```python
# lmplot ads two regression lines
plt.figure(figsize=(15,6))
sns.lmplot(x="age", y= "total_claim_amount", hue="fraud_reported", data=insurance_data)
```

```python
# a SWARMPLOT used a categorical variable in the plot
# it helps to see the density of points in each category

sns.swarmplot(x="fraud_reported", y= "total_claim_amount", data=insurance_data)
```
