```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Simulated data — e.g., exam scores or ages
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=200)  # mean=70, std=10

# 1. Just histogram
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.histplot(data, bins=20, color='skyblue')
plt.title("Histogram")

# 2. Just KDE
plt.subplot(1, 3, 2)
sns.kdeplot(data, fill=True, color='orange')
plt.title("KDE Plot")

# 3. Histogram + KDE overlay
plt.subplot(1, 3, 3)
sns.histplot(data, bins=20, color='skyblue', kde=True)
plt.title("Histogram + KDE")

plt.tight_layout()
plt.show()
```

```python
# Create a DataFrame for compatibility with Seaborn
import pandas as pd
df = pd.DataFrame({'scores': data})


# Set up the plots
plt.figure(figsize=(12, 4))

# 1. Box Plot
plt.subplot(1, 2, 1)
sns.boxplot(x='scores', data=df, color='lightgreen')
plt.title("Box Plot")

# 2. Violin Plot
plt.subplot(1, 2, 2)
sns.violinplot(x='scores', data=df, inner='box', color='orchid')
plt.title("Violin Plot")

plt.tight_layout()
plt.show()
```
