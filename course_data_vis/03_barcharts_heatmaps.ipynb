```python
# Environment setup
import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

filepath = "../data/flight_delays.csv"
flights = pd.read_csv(filepath, index_col="Month")


# print all flights
flights
```

```python
# plots a bar chart showing the average arrival delay for NK airlines
plt.figure(figsize=(10, 6))

# add title
plt.title("Average Arrival Delay for NK airlines Flights, by Month")

sns.barplot(x=flights.index, y=flights["NK"]) 
# NOTE: We always have to use ds.INDEX to index the rows

plt.ylabel("Arrival delay in minutes")
```

```python
# heatmaps
plt.figure(figsize=(18, 16))

plt.title("Average Arrival Delay for Each airline, by Month")

sns.heatmap(data=flights, annot=True)
plt.xlabel("Airline")
```
